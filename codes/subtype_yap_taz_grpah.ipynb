```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = r"E:\iiser_project\result_broad_filtered.xls"
excel_data = pd.ExcelFile(file_path)

# Iterate through each sheet
for sheet_name in excel_data.sheet_names:
    # Load the sheet into a DataFrame
    df = excel_data.parse("yap1_wwtr1_counts_details1")

    # Standardize column names by stripping extra spaces
    df.columns = df.columns.str.strip()
df.head(27)
```

```python
df_filtered = df[df['Patient'] != 'Total']
df_filtered.head(27)
```

```python
import pandas as pd

# Create a DataFrame for the subtype information
subtype_data = {
    "Patient": [
        "CID3586", "CID3838", "CID3921", "CID3941", "CID3946", "CID3948", 
        "CID3963", "CID4040", "CID4066", "CID4067", "CID4290A", "CID4398",
        "CID44041", "CID4461", "CID4463", "CID4465", "CID4471", "CID4495",
        "CID44971", "CID44991", "CID4513", "CID4515", "CID45171", "CID4523",
        "CID4530N", "CID4535"
    ],
    "Subtype": [
        "HER2+", "HER2+", "HER2+", "ER+", "TNBC", "ER+", 
        "TNBC", "ER+", "HER2+", "ER+", "ER+", "ER+",
        "TNBC", "ER+", "ER+", "TNBC", "ER+", "TNBC",
        "TNBC", "TNBC", "TNBC", "TNBC", "HER2+", "TNBC",
        "ER+", "ER+"
    ]
}
# Convert to a DataFrame
subtype_df = pd.DataFrame(subtype_data)

# Merge with the gene counts dataset
df_merged = pd.merge(df_filtered, subtype_df, on='Patient')

# Display the merged DataFrame
df_merged.head(28)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the columns for analysis
columns_of_interest = [
    'Total_YAP1_Cell_Count', 
    'Total_WWTR1_Cell_Count', 
    'Both_YAP1_WWTR1_Cell_Count'
]

# Group data by subtype and compute mean or sum
grouped_data = df_merged.groupby('Subtype')[columns_of_interest].sum()

# Prepare data for plotting
plot_data = grouped_data.reset_index().melt(id_vars='Subtype', var_name='Gene Expression', value_name='Cell Count')

# Define a new vibrant color palette
custom_palette = {
    'Total_YAP1_Cell_Count': '#FF5733',  # Bright orange-red
    'Total_WWTR1_Cell_Count': 'purple',  # Vibrant green
    'Both_YAP1_WWTR1_Cell_Count': '#3357FF'  # Bright blue
}

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(
    data=plot_data,
    x='Subtype',
    y='Cell Count',
    hue='Gene Expression',
    palette=custom_palette
)

# Add numbers in a box on top of each bar
for i, bar in enumerate(plt.gca().patches):
    # Retrieve the height (value) of each bar
    value = bar.get_height()
    if value > 0:  # Only label bars with positive values
        plt.text(
            bar.get_x() + bar.get_width() / 2,  # Center the text horizontally
            bar.get_height() + max(plot_data['Cell Count']) * 0.01,  # Slightly above the bar
            f"{int(value)}", 
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'),
            color='black',
            ha='center',
            va='bottom'
        )

# Customize plot
plt.title('YAP1, WWTR1 (TAZ), and Co-expression Counts by Subtype', fontsize=16)
plt.xlabel('Subtype', fontsize=14)
plt.ylabel('Cell Count', fontsize=14)
plt.legend(title='Expression Type', loc='upper left')
plt.tight_layout()

# Show the plot
plt.savefig('subtype_yap_wwtr1.png')
plt.show()
```

```python
# Debugging: Print column names
print("Columns in DataFrame:", df_merged.columns)

# Check for 'Subtype' column existence
if 'Subtype' not in df_merged.columns:
    print("The 'Subtype' column is missing. Please verify the DataFrame structure.")
else:
    # Define colors for each subtype
    subtype_colors = {
        'HER2+': '#FF5733',  # Bright red for HER2+
        'ER+': '#33FF57',    # Bright green for ER+
        'TNBC': '#3357FF'    # Bright blue for TNBC
    }

    # Plot Percentages
    plt.figure(figsize=(14, 8))

    for subtype, data in df_merged.groupby('Subtype'):
        # YAP1 percentage
        plt.plot(
            data['Patient'], 
            data['YAP_percentage'], 
            marker='o',  # Circle marker
            linestyle='-',  # Solid line
            label=f'YAP ({subtype})',
            color=subtype_colors[subtype]
        )

        # WWTR1 percentage
        plt.plot(
            data['Patient'], 
            data['WWTR1_percentage'], 
            marker='s',  # Square marker
            linestyle='--',  # Dashed line
            label=f'WWTR1 ({subtype})',
            color=subtype_colors[subtype]
        )

        # Both YAP1 and WWTR1 percentage
        plt.plot(
            data['Patient'], 
            data['both_YAP_TAZ_percentage'], 
            marker='^',  # Triangle marker
            linestyle=':',  # Dotted line
            label=f'Both ({subtype})',
            color=subtype_colors[subtype]
        )

    # Customize the plot
    plt.title('Percentages of YAP1, WWTR1, and Both per Patient by Subtype', fontsize=16)
    plt.ylabel('Percentage (%)', fontsize=14)
    plt.xlabel('Patient', fontsize=14)

    # Ensure all patient IDs are shown
    plt.xticks(range(len(df_merged['Patient'])), df_merged['Patient'], rotation=45, fontsize=12)

    # Adjust legend
    plt.legend(
        title='Subtype and Gene Expression', 
        fontsize=8, 
        title_fontsize=12, 
        loc='upper left'
    )
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Save and Display the plot
    plt.savefig("percentage_line_with_subtype_differentiated.png")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the columns for analysis
columns_of_interest = [
    'Total_YAP1_Cell_Count', 
    'Total_WWTR1_Cell_Count', 
    'Both_YAP1_WWTR1_Cell_Count'
]

# Group data by subtype and compute mean
grouped_data = df_merged.groupby('Subtype')[columns_of_interest].mean()

# Prepare data for plotting
plot_data = grouped_data.reset_index().melt(id_vars='Subtype', var_name='Gene Expression', value_name='Cell Count')

# Define a new vibrant color palette
custom_palette = {
    'Total_YAP1_Cell_Count': '#FF5733',  # Bright orange-red
    'Total_WWTR1_Cell_Count': 'purple',  # Vibrant green
    'Both_YAP1_WWTR1_Cell_Count': '#3357FF'  # Bright blue
}

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(
    data=plot_data,
    x='Subtype',
    y='Cell Count',
    hue='Gene Expression',
    palette=custom_palette
)

# Add numbers in a box on top of each bar
for i, bar in enumerate(plt.gca().patches):
    # Retrieve the height (value) of each bar
    value = bar.get_height()
    if value > 0:  # Only label bars with positive values
        plt.text(
            bar.get_x() + bar.get_width() / 2,  # Center the text horizontally
            bar.get_height() + max(plot_data['Cell Count']) * 0.01,  # Slightly above the bar
            f"{value:.2f}",  # Show up to two decimal places for the average
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'),
            color='black',
            ha='center',
            va='bottom'
        )

# Customize plot
plt.title('Average YAP1, WWTR1 (TAZ), and Co-expression Counts by Subtype', fontsize=16)
plt.xlabel('Subtype', fontsize=14)
plt.ylabel('Average Cell Count', fontsize=14)
plt.legend(title='Expression Type', loc='upper left')
plt.tight_layout()

# Show the plot
plt.savefig('average_subtype_yap_wwtr1.png')
plt.show()
```
